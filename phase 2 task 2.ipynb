```python
import pandas as pd 
```

```python
url = "dataset - netflix12.csv"
```

```python
dataset = pd.read_csv(url)
```

```python
dataset.head()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8, 6))
```

```python
plt.hist(dataset['release_year'], bins=10, color='skyblue', edgecolor='black')
plt.ylabel('Y-Axis Label')
plt.xlabel('X-Axis Label') 
plt.title('Histogram')
plt.show()
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='release_year', data=dataset, color='red')
plt.xlabel('X-Axis Label')
plt.ylabel('Y-Axis Label')
plt.title('Box Plot')
plt.show()
```

```python
plt.figure(figsize=(3,5))
plt.scatter(dataset['type'], dataset['release_year'], color='green')
plt.xlabel('X-Axis Label')
plt.ylabel('Y-Axis Label')
plt.title('Scatter Plot')
plt.show()
```
