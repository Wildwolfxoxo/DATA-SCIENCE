```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
train=pd.read_csv('train.csv')
test=pd.read_csv('test.csv')
```

```python
train.head()
```

```python
test.head()
```

```python
train.isnull().sum()
```

```python
train.dropna(inplace=True)
```

```python
train.isnull().sum()
```

```python
test.isnull().sum()
```

```python
x_train=np.array(train['x']).reshape(-1,1)
y_train=np.array(train['y']).reshape(-1,1)
x_test=np.array(test['x']).reshape(-1,1)
y_test=np.array(test['y']).reshape(-1,1)
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
predicted=model.predict(x_test)
```

```python
predicted
```

```python
plt.scatter(x_train, y_train, color='orange', label='Train Data')
plt.scatter(x_test, y_test, color='green', label='Test Data')
plt.plot(x_test, predicted, color='blue', label='Regression Line')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Regression')
plt.legend()
plt.show()
```
